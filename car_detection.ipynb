```python
# Step 1: Install required libraries
!pip install ultralytics opencv-python-headless pandas

# Step 2: Import necessary libraries
import cv2
import pandas as pd
from ultralytics import YOLO
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files

# Step 3: Load the YOLOv8 model
model = YOLO('yolov8m.pt')  # You can switch to a smaller model like yolov8s.pt for faster processing

# Step 4: Upload your video file
#uploaded = files.upload()

# Step 5: Set up video capture
video_path = "/content/4K.mp4"  # Get the uploaded video file path
cap = cv2.VideoCapture(video_path)

frame_width, frame_height = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
frame_rate = cap.get(cv2.CAP_PROP_FPS)
output_path = "/content/drive/MyDrive/car detection_countoutput_car_count.mp4"
out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*"mp4v"), frame_rate, (frame_width, frame_height))

# Initialize the MultiTracker from OpenCV
multi_tracker = cv2.legacy.MultiTracker_create()

# Step 6: Process video frames and count cars
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("End of video or error reading frame.")
        break

    # YOLO car detection
    results = model.predict(frame)
    detections = results[0].boxes.data  # Extract detections

    # If no detections, continue to next frame
    if len(detections) == 0:
        out.write(frame)  # Write the frame even if no detection occurs
        continue

    # Convert tensor to NumPy array and DataFrame
    detections = detections.cpu().numpy()  # Move to CPU and convert to NumPy
    df = pd.DataFrame(detections).astype(float)

    car_list = []

    # Iterate over the detections and extract only cars
    for _, row in df.iterrows():
        x1, y1, x2, y2, confidence, class_id = int(row[0]), int(row[1]), int(row[2]), int(row[3]), row[4], int(row[5])

        # Filter out only cars (YOLO class id for 'car' is typically 2)
        if class_id == 2:  # Class 2 corresponds to 'car' in YOLOv8
            car_list.append([x1, y1, x2, y2])

    # If it's the first frame or no cars are being tracked, initialize trackers
    if len(multi_tracker.getObjects()) == 0:
        for bbox in car_list:
            tracker = cv2.legacy.TrackerKCF_create()  # Using KCF tracker for each car
            multi_tracker.add(tracker, frame, tuple(bbox))  # Add to MultiTracker

    # Update trackers
    success, boxes = multi_tracker.update(frame)

    # Draw bounding boxes and car count
    car_count = 0
    for i, newbox in enumerate(boxes):
        p1 = (int(newbox[0]), int(newbox[1]))
        p2 = (int(newbox[0] + newbox[2]), int(newbox[1] + newbox[3]))
        cv2.rectangle(frame, p1, p2, (0, 255, 0), 2, 1)
        car_count += 1

    # Display car count on the screen
    cv2.putText(frame, f'Cars on Screen: {car_count}', (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Write frame to output video
    out.write(frame)

# Step 7: Release video resources
cap.release()
out.release()

# Step 8: Display the output video file
print("Car detection and counting finished. Download the video:")

# Download the processed video
files.download(output_path)
```
